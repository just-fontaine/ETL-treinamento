```python
'''

EXPLICAÇÃO DOS COMANDOS

left_on     chave da tabela a esquerda
right_on    chave da tabela a direita
how         tipo de join


TIPOS DE JOIN 

left          todos os dados da tabela a esquerda (1° tabela) e completa com o restante
right         todos os dados da tabela a direita (2° tabela) e completa com o restante
inner         retorna dados apenas quando as duas tabelas tem chaves  
outer         retorna todos os registros de ambas as tabelas

'''
```
