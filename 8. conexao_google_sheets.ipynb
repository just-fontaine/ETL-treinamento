```python
import sys
import subprocess

def instalar_e_importar(nome_biblioteca):
    try:
        # Tenta importar a biblioteca
        return __import__(nome_biblioteca)
    except ImportError:
        print(f"A biblioteca '{nome_biblioteca}' não está instalada. Instalando agora...")
        try:
            # Executa o comando pip para instalar a biblioteca
            subprocess.check_call([sys.executable, "-m", "pip", "install", nome_biblioteca])
            # Tenta importar novamente após a instalação
            return __import__(nome_biblioteca)
        except subprocess.CalledProcessError:
            print(f"Falha ao instalar a biblioteca '{nome_biblioteca}'. Verifique sua conexão com a internet e as permissões.")
            sys.exit(1)
```
