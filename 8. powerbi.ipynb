```python
import pandas as pd
import os
import subprocess
import sys

def install_matplotlib():
    subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib"])

# Tente importar o matplotlib, e se nÃ£o estiver instalado, instale-o.
try:
    import matplotlib
except ModuleNotFoundError:
    install_matplotlib()
    import matplotlib

# DIRÃ‰TORIO

diretorio = (r"C:\fora do onedrive\Python Scripts\Arquivos\duas metades")

# LISTA PARA ARMAZENAR OS DADOS DOS ARQUIVOS

dados = []

for arquivo in os.listdir(diretorio):
    if arquivo.endswith('csv'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_csv(caminho_arquivo)
        dados.append(df)

# CONSOLIDAR EM UM UNICO DATA FRAME

consolidado = pd.concat(dados)
consolidado
```
