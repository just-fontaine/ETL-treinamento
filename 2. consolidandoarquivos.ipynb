##### CONSOLIDANDO ARQUIVOS

```python
# DIRÉTORIO

diretorio = (r"C:\fora do onedrive\Python Scripts\Arquivos\vendas")

# LISTA PARA ARMAZENAR OS DADOS DOS ARQUIVOS

dados = []

for arquivo in os.listdir(diretorio):
    if arquivo.endswith('csv'):
        caminho_arquivo = os.path.join(diretorio, arquivo)
        df = pd.read_csv(caminho_arquivo)
        dados.append(df)

# CONSOLIDAR EM UM UNICO DATA FRAME

vendas_2010_2018 = pd.concat(dados)

vendas_2010_2018
```

```python
pasta_destino = r'C:\fora do onedrive\Python Scripts\Arquivos\vendas\Arquivos Consolidados'

nome_arquivo = 'vendas_2010_2018.txt'

caminho_completo = pasta_destino + nome_arquivo

# SUBSTITUIR 

if os.path.exists(caminho_completo):
    os.remove(caminho_completo)

vendas_2010_2018.to_csv(caminho_completo, index=False, sep=(','))
```
